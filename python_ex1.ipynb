### 3. A의 주민등록번호는 901115-1234567이다. A의 주민등록번호를 연월일(yymmdd)부분과 그 뒤의 숫자부분으로 나누어 출력해보자.

```python
pin = "901115-1234567"
yymmdd = pin[0:6]
num = pin[-7:]
print("주민등록 앞자리:", yymmdd)
print("주민등록 뒷자리:", num)
```
